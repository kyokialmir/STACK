```python
class Movie:
    def __init__(self, title, genre, release_year):
        self.title = title
        self.genre = genre
        self.release_year = release_year

    def __str__(self):
        return f"{self.title} ({self.release_year}) - {self.genre}"

class MovieDatabase:
    def __init__(self):
        self.stack = []

    def add_movie(self, movie):
        self.stack.append(movie)

    def remove_movie(self):
        if self.stack:
            return self.stack.pop()
        else:
            print("No movies to remove.")
            return None

    def print_all_movies(self):
        for movie in reversed(self.stack):
            print(movie)

    def search_movie_by_title(self, title):
        for movie in reversed(self.stack):
            if movie.title.lower() == title.lower():
                print(movie)
                return
        print(f"No movie found with title '{title}'.")

    def list_movies_by_genre(self, genre):
        found = False
        for movie in reversed(self.stack):
            if movie.genre.lower() == genre.lower():
                print(movie)
                found = True
        if not found:
            print(f"No movies found in genre '{genre}'.")

    def list_movies_by_release_year(self, release_year):
        found = False
        for movie in reversed(self.stack):
            if movie.release_year == release_year:
                print(movie)
                found = True
        if not found:
            print(f"No movies found from the year '{release_year}'.")

# Пример использования базы данных фильмов
if __name__ == "__main__":
    db = MovieDatabase()

    while True:
        print("\n1. Add movie\n2. Remove movie\n3. Print all movies\n4. Search movie by title\n5. List movies by genre\n6. List movies by release year\n7. Exit")
        choice = int(input("Choose an option: "))

        if choice == 1:
            title = input("Enter movie title: ")
            genre = input("Enter movie genre: ")
            release_year = int(input("Enter release year: "))
            movie = Movie(title, genre, release_year)
            db.add_movie(movie)
        elif choice == 2:
            removed_movie = db.remove_movie()
            if removed_movie:
                print(f"Removed movie: {removed_movie}")
        elif choice == 3:
            db.print_all_movies()
        elif choice == 4:
            title = input("Enter movie title to search: ")
            db.search_movie_by_title(title)
        elif choice == 5:
            genre = input("Enter genre to list movies: ")
            db.list_movies_by_genre(genre)
        elif choice == 6:
            release_year = int(input("Enter release year to list movies: "))
            db.list_movies_by_release_year(release_year)
        elif choice == 7:
            break
        else:
            print("Invalid choice, please try again.")
```
